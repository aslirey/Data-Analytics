```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plotdata = pd.DataFrame({
    "Charles":[8.5,0.5,3.0,1.0,1.5,3.0,6.5],
    "Henry":[9.5,1.0,2.0,1.5,0.5,2.5,7.0],
    "Susan":[7.0,1.5,1.0,2.5,2.0,2.0,8.0]
    }, 
    index=['Work', 'Family', 'Homeworks', 'Individual', 'Socializing','Spare Time','Sleep']
)
```

```python
plotdata.plot(kind="bar")
plt.title("How many hours spend in daily activities?")
plt.xlabel("Area of Interest")
plt.ylabel("Hours")
```
